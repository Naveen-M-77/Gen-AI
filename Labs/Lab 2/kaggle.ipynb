```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_df = pd.read_csv("/kaggle/input/mnist-in-csv/mnist_train.csv")
test_df = pd.read_csv("/kaggle/input/mnist-in-csv/mnist_test.csv")
```

```python
print("Training data shape:", train_df.shape)
print("Testing data shape:", test_df.shape)
```

```python
X = train_df.drop("label", axis=1)
y = train_df["label"]
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x=y)
plt.title("Distribution of Digits in MNIST Dataset")
plt.xlabel("Digit Label")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,5))

for i in range(5):
    image = X.iloc[i].values.reshape(28,28)
    label = y.iloc[i]

    plt.subplot(1,5,i+1)
    plt.imshow(image, cmap='gray')
    plt.title(f"Label: {label}")
    plt.axis('off')

plt.show()
```

```python
plt.hist(X.iloc[0], bins=30)
plt.title("Pixel Intensity Distribution of a Sample Image")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.show()
```

```python
mean_image = X.mean().values.reshape(28,28)

plt.imshow(mean_image, cmap='gray')
plt.title("Average Image of MNIST Dataset")
plt.axis('off')
plt.show()
```
